```python
for fi, ti, ft, tt in zip(fixed_input_dir.iterdir(), temporal_input_dir.iterdir(), fixed_target_dir.iterdir(), temporal_target_dir.iterdir()):
    fi = np.load(fi).squeeze()
    ti = np.load(ti).squeeze()
    ft = np.load(ft).squeeze()
    tt = np.load(tt).squeeze()

    subgrid_orog = ft[:, :, 1]
    standard_deviation_of_orog = np.std(subgrid_orog, axis=1)
    standard_deviation_of_orog = standard_deviation_of_orog.reshape(12,12)
    plt.pcolormesh(standard_deviation_of_orog)
    plt.colorbar()
    plt.title("Standard deviation of subgrid orography")
    break
```

```python
vertical_levels = np.arange(70)
def normalise(profile, variable):
    if variable == "specific_humidity":
        return profile * 0.025
    if variable == "pressure":
        return profile * 106000
    if variable == "temperature":
        return profile * 180 + 140
    raise ValueError(f"Not a valid variable name: {variable}")
def get_units(variable):
    units_dict = {
        "specific_humidity": "kg/kg",
        "pressure": "Pa",
        "temperature": "K",
    }
    return units_dict[variable]

for fi, ti, ft, tt in zip(fixed_input_dir.iterdir(), temporal_input_dir.iterdir(), fixed_target_dir.iterdir(), temporal_target_dir.iterdir()):
    fi = np.load(fi).squeeze()
    ti = np.load(ti).squeeze()
    ft = np.load(ft).squeeze()
    tt = np.load(tt).squeeze()
    for index, variable in zip([0,1,2], ["specific_humidity", "pressure", "temperature"]):
        profile = normalise(ti[0, :, index], variable)
        subgrid_profiles = normalise(tt[0, :, :, index], variable)
        for i in range(100):
            subgrid_anomaly = (subgrid_profiles[:, i] - profile).squeeze()
            plt.plot(subgrid_anomaly, vertical_levels)
        plt.xlabel(f"{variable} anomaly ({get_units(variable)})")
        plt.ylabel("Vertical levels")
        plt.show()

    break

```

NB: Would be good to remove the vertical variation by normalising by the means and standard deviations of
the variables at each vertical level (or some other normalisation method, just something to get everything
onto the same scale)

BUT it's also important to consider that for example humidity high up is basically zero so we really care a lot more about the humidity at low altitudes, so any normalisation method that removes this distinction is probably not so good. Perhaps applying some simple linear weighting function above a certain altitude would be good for humidity.

As above, learning anomalies is probably better than learning the full profiles because the subgrid profiles are always basically the mean plus some small perturbation.

# Avoiding negative humidities

If the anomaly to the humidity is learned, it's very important that the mean + anomaly is not negative.
I suppose you could solve that by clamping the anomaly, not sure how to deal with it otherwise.

Probably don't need to worry about the temperature and pressures going negative because the anomalies are
so much smaller than the means.
